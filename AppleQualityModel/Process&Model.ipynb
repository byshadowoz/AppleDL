```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = pd.read_csv('apple_quality.csv', index_col=0)
data
```

```python
data.info()
```

```python
data.dropna()
```

```python
data['Acidity'] = data['Acidity'].astype(float)
```

```python
data['Quality'] = data['Quality'].map({'good': 1, 'bad': 0}).astype(int)
```

```python
data.info()
```

```python
model = Sequential()
model.add(Dense(12, input_dim=7, activation='LeakyReLU'))#entrada/oculta layer
model.add(Dense(8, activation='LeakyReLU'))#oculta layer
model.add(Dense(1, activation='sigmoid'))#salida layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X = data.drop('Quality', axis=1)
Y = data.Quality
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
model.fit(X_train, y_train, validation_split= 0.1, epochs=150, batch_size=10)
```

```python
print('evaluate model against trained data')
score = model.evaluate(X_train, y_train, verbose=0)
print('score %s: %.2f%%' % (model.metrics_names[1], score[1]*100))

print('evaluate model against new data')
score = model.evaluate(X_test, y_test, verbose=0)
print('score %s: %.2f%%' % (model.metrics_names[1], score[1]*100))
```

```python
trained_model_file = 'trained_Apple_Quaity_model'
model.save(trained_model_file)
print('saved traines model to disk :', trained_model_file)
```
